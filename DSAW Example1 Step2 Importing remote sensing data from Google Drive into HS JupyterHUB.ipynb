## Step 1 Configure HS to run Google Drive API.

```python
# from __future__ import print_function
# from googleapiclient.discovery import build
# from httplib2 import Http
# from oauth2client import file, client, tools
import io
from googleapiclient.http import MediaIoBaseDownload
from apiclient import discovery, errors
from httplib2 import Http
from oauth2client import client, file, tools

# Initializing display
from IPython.display import Image
%matplotlib inline
```

```python
# define variables
credentials_file_path = './credentials/credentials.json'
clientsecret_file_path = './credentials/client_secret.json'

# verifying the Google Drive API is up and running
SCOPES = 'https://www.googleapis.com/auth/drive'

# define store
store = file.Storage(credentials_file_path)
credentials = store.get()

if not credentials or credentials.invalid:
    flow = client.flow_from_clientsecrets(clientsecret_file_path, SCOPE)
    credentials = tools.run_flow(flow, store)
    
# define API service
http = credentials.authorize(Http())
service = discovery.build('drive', 'v3', http=http)# define API service
http = credentials.authorize(Http())
service = discovery.build('drive', 'v3', http=http)
```

We need the name of the files exported from Earth Engine. In this case landsat7_2012-06-30.tif

```python
# NOTE: maybe there is a way to pass a query into the service, this is an attempt (did not work)
# filename='LC08_123032_20140515_TOA.tif' #here we identify the file of interest
# query='\"name contains \''+image_name+'\'\"'
# print(query)
```

```python
# Call the Drive v3 API
results = service.files().list(q="name contains 'landsat7_'",pageSize=100, fields="nextPageToken, files(name, id)").execute()
# results = service.files().list(q=print(query),pageSize=100, fields="nextPageToken, files(name, id)").execute()

items = results.get('files', [])

import pandas as pd
df =pd.DataFrame.from_dict(items)
print(df)
```

The previous cell should give an id like this:*xogAczbNGN8Em_WH6NH8v5GiRSuFMfnV_-SX5oJyxww*for every image we downloaded from GEE. This is an internal id that indicates that the file is in fact within Google Drive

For GoogleDrive API, we need the id of the file to download to HydroShare

```python
id = df.iloc[0:4]['id']
print(id)
```

```python
print(id[3])
```

the following cell will download all our files.

```python
i=0
for file in id:

    file_id = file
    request = service.files().get_media(fileId=file_id)
    fh = io.BytesIO()
    downloader = MediaIoBaseDownload(fh, request)
    done = False
    while done is False:
        status, done = downloader.next_chunk()
        print ("Download %d%%." % int(status.progress() * 100))
# The file is downloaded to memory. It needs to be written to a file.
    #writting the file from memory to local storage
    fh.seek(0)

    with open(df.iloc[i]['name'], "wb") as wer:
        wer.write(fh.read())    
    i=i+1    
```

That's it! The files are available for any other calculation in HS

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

from rasterio.plot import show

fp1 = './landsat7_2012-05-13.tif'
fp2 = './landsat7_2012-05-29.tif'
fp3 = './landsat7_2012-06-14.tif'
fp4 = './landsat7_2012-06-30.tif'

def im_vis(fp):
    # Open the file:
    raster = rasterio.open(fp)

    # Convert to numpy arrays
    nir = raster.read(4)
    red = raster.read(3)
    green = raster.read(2)

    # Stack bands
    nrg = np.dstack((nir*0.0001, red*0.0001, green*0.0001))
    return nrg

image1=im_vis(fp1)
image2=im_vis(fp2)
image3=im_vis(fp3)
image4=im_vis(fp4)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4)

ax1.imshow(image1)
ax1.axis('off')
ax2.imshow(image2)
ax2.axis('off')
ax3.imshow(image3)
ax3.axis('off')
ax4.imshow(image4)
ax4.axis('off')
```
